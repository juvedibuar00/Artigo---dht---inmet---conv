```python
# ==============================================
# 1. IMPORTAÇÕES
# ==============================================
import pandas as pd
import io
from google.colab import files
import numpy as np

# ==============================================
# 2. UPLOAD DO ARQUIVO (use o horas.csv)
# ==============================================
print("📤 Faça o upload do arquivo 'horas.csv'")
uploaded = files.upload()
filename = list(uploaded.keys())[0]

# Lê o CSV
df = pd.read_csv(io.BytesIO(uploaded[filename]), encoding='utf-8')

# Remove espaços extras dos nomes das colunas
df.columns = df.columns.str.strip()

# ==============================================
# 3. DIAGNÓSTICO
# ==============================================
print("\n🔍 Nomes das colunas:")
print(df.columns.tolist())
print("\n🔍 Primeiras 5 linhas:")
print(df.head())
print("\n🔍 Total de linhas:", len(df))

# ==============================================
# 4. CRIAR UMA COLUNA 'Dia' SEQUENCIAL
# ==============================================
# Cada dia tem 24 horas (0 a 23)
# Vamos assumir que os dados estão perfeitamente ordenados e completos
df['Dia'] = df.index // 24 + 1  # Dia 1, 2, 3, ...

# ==============================================
# 5. FILTRAR HORÁRIOS 9, 15, 21
# ==============================================
horarios = [9, 15, 21]
df_filtrado = df[df['Hora'].isin(horarios)].copy()

print(f"\n📊 Total de linhas após o filtro: {len(df_filtrado)}")
if len(df_filtrado) == 0:
    print("❌ Nenhuma linha encontrada para os horários 9, 15, 21.")
    raise SystemExit("Processo interrompido.")

# ==============================================
# 6. PIVOTEAR – cada dia vira uma linha
# ==============================================
pivot = df_filtrado.pivot_table(
    index='Dia',
    columns='Hora',
    values=['Temp_ DHT1', 'UR_DHT11', 'Temp_INMET', 'UR_INMET']
)

# Achatar as colunas
pivot.columns = [f'{col[0]}_{col[1]}h' for col in pivot.columns]

# Renomear para nomes claros
rename_map = {
    'Temp_ DHT1_9h': 'DHT11_09h_Temp',
    'UR_DHT11_9h': 'DHT11_09h_UR',
    'Temp_ DHT1_15h': 'DHT11_15h_Temp',
    'UR_DHT11_15h': 'DHT11_15h_UR',
    'Temp_ DHT1_21h': 'DHT11_21h_Temp',
    'UR_DHT11_21h': 'DHT11_21h_UR',
    'Temp_INMET_9h': 'INMET_09h_Temp',
    'UR_INMET_9h': 'INMET_09h_UR',
    'Temp_INMET_15h': 'INMET_15h_Temp',
    'UR_INMET_15h': 'INMET_15h_UR',
    'Temp_INMET_21h': 'INMET_21h_Temp',
    'UR_INMET_21h': 'INMET_21h_UR',
}
pivot.rename(columns=rename_map, inplace=True)

# Resetar índice para ter 'Dia' como coluna
pivot.reset_index(inplace=True)

# ==============================================
# 7. EXIBIR PRIMEIRAS LINHAS
# ==============================================
print("\n✅ Tabela reorganizada (primeiras 5 linhas):")
print(pivot.head())

# ==============================================
# 8. SALVAR CSV
# ==============================================
csv_file = 'horarios_9_15_21_por_dia.csv'
pivot.to_csv(csv_file, index=False, encoding='utf-8-sig')
print(f"\n💾 CSV salvo como '{csv_file}'")

# ==============================================
# 9. SALVAR EXCEL com multi-índice (opcional)
# ==============================================
colunas_multi = pd.MultiIndex.from_tuples([
    ('DHT11', '09h', 'Temp'),
    ('DHT11', '09h', 'UR'),
    ('DHT11', '15h', 'Temp'),
    ('DHT11', '15h', 'UR'),
    ('DHT11', '21h', 'Temp'),
    ('DHT11', '21h', 'UR'),
    ('INMET', '09h', 'Temp'),
    ('INMET', '09h', 'UR'),
    ('INMET', '15h', 'Temp'),
    ('INMET', '15h', 'UR'),
    ('INMET', '21h', 'Temp'),
    ('INMET', '21h', 'UR'),
])
df_excel = pivot.set_index('Dia')
df_excel.columns = colunas_multi
excel_file = 'horarios_9_15_21_por_dia.xlsx'
df_excel.to_excel(excel_file, index=True, header=True)
print(f"💾 Excel salvo como '{excel_file}'")

# ==============================================
# 10. DOWNLOAD
# ==============================================
files.download(csv_file)
files.download(excel_file)

print("\n🎉 Downloads iniciados! Verifique sua pasta de downloads.")
```

```python
# Lê o arquivo mantendo o cabeçalho original
df = pd.read_csv('dados_oficial_horario.csv', header=0)

# A primeira coluna (sem nome) é a data – vamos renomeá-la
df.rename(columns={df.columns[0]: 'Data'}, inplace=True)

# Visualizar as primeiras linhas
df.head()
```

```python
# Substitui strings vazias por NaN e faz o preenchimento para frente
df['Data'] = df['Data'].replace('', np.nan).ffill()

# Agora todas as linhas têm uma data associada
```

```python
# Converter a hora para inteiro (caso esteja como string)
df['Horas'] = df['Horas'].astype(int)

# Criar uma coluna com data+hora
df['Datetime'] = pd.to_datetime(df['Data'] + ' ' + df['Horas'].astype(str) + ':00', format='%d-%b-%y %H:%M')
# Nota: o mês está em inglês (Oct, Nov, etc.) – o formato '%d-%b-%y' reconhece isso.
```

```python
# Extrair a data (sem hora) para agrupamento
df['Data_somente'] = df['Datetime'].dt.date

# Selecionar apenas as colunas de interesse (temperaturas e umidades)
colunas_medias = ['TEMP_DHT11', 'UR DHT11', 'TEMP_INMET', 'UR INMEMT']

# Agrupar e calcular a média
medias_diarias = df.groupby('Data_somente')[colunas_medias].mean().reset_index()

# Renomear a coluna de data
medias_diarias.rename(columns={'Data_somente': 'Data'}, inplace=True)

# Visualizar o resultado
medias_diarias.head()
```

```python
# Salvar como arquivo Excel
medias_diarias.to_excel('medias_diarias.xlsx', index=False, sheet_name='Medias Diarias')

# Baixar o arquivo para o computador
files.download('medias_diarias.xlsx')
```
